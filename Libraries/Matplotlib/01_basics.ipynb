```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y=[10,20,25,30]
```

```python
plt.plot(x,y)
plt.show()
```

### LINE PLOT

```python
plt.plot([1,2,3],[10,20,30],label="Trend",marker="o",linestyle="--", color="orange")
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axix")
plt.show()
```

### BAR PLOT

```python
categories = ["A","B","C"]
values = [10,15,7]
plt.bar(categories, values, color="blue")
plt.title("Bar Chart")
plt.show(
    
)
```

#### Histogram

```python
data = [1,2,2,3,3,3,4,4,4,4]
plt.hist(data,bins=4,color="green",edgecolor="black")
plt.title("Histogram")
plt.show()
```

#### Scatter PLot

```python
plt.scatter(x,y,color="red")
plt.title("Scatter Plot")
plt.show()
```
